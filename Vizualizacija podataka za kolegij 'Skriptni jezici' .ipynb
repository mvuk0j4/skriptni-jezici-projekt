```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import requests
import json 
import matplotlib.patches as mpatches
```

```python
def get_timeseries_data(group_id: int):
    r = requests.get(f"https://apikomore.mci.ba/fakultet/graph_data/?start=2022-05-01&end=2022-05-31&group_id={group_id}&apikey=skriptni-jezici-2022") 
    res = json.loads(r.text)
    return res
```

```python
def visualize_data(group_id):
    res = get_timeseries_data(group_id)
    df = pd.DataFrame(res['temperatures'])
    ax, ay  = [],[]
    df["temp"] = df['temp'].astype(int)
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    ay = df['temp']
    ax = df['timestamp']
    plt.style.use('seaborn')
    df.plot(x='timestamp', y='temp', kind='line',color='blue', figsize=(20,8))
    # plt.plot(ax , ay, color='#FF33E9',linewidth = '1')
    plt.axhline(res['set_temp'], color='red',ls='-', linewidth=0.7)
    plt.xlabel('Vrijeme', fontdict={'fontsize': 15})
    plt.ylabel('Temperatura', fontdict={'fontsize': 15})
    plt.title(res["name"],fontdict={'fontsize': 30})
    set_temp_legend = mpatches.Patch(color='red', label='Zadana temperatura')
    temp_legend = mpatches.Patch(color='blue', label='Temperatura')
    plt.legend(handles=[set_temp_legend, temp_legend])
    plt.show()
```
